```python
"""* Write a program to check palindrome."""

inputstring = input("Enter the String: ")

reversestring = inputstring[::-1]

if inputstring == reversestring:
  print("It is a palindrome.")
else:
  print("It is not a palindrome.")


```

```python
"""*  Write a program to count no. of vowels and print all the vowels."""


vowels = ["a","e","i","o","u"]
countvalue = 0
resultstring = ""

inputvalue = input("Enter the value: ")

for vow in vowels:
  if vow in inputvalue:
    countvalue += inputvalue.count(vow)
    if not resultstring:
      resultstring = vow
    else:
      resultstring = resultstring+","+vow
  else:
    continue

print("no.of vowles in the string: "+str(countvalue))
print("Vowels present in the string are: "+resultstring)







```
